```python
# Import necessary libraries
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
import sklearn.model_selection as skm
```

```python
def b1(x):
    if x < 0:
        return 0
    elif 0<= x <1:
        return 1
    elif 1<= x <2:
        return 2-x
    else:
        return 0
```

```python
# Visualization of the function
fig, ax = subplots(figsize=(10, 6))
x_plot = np.linspace(-2, 2, 1000)
y_plot = [1 + 1 * b1(x) for x in x_plot]

# Plot the function
ax.plot(x_plot, y_plot, 'b-', linewidth=2, label='y = 1 + b₁(x)')

# Highlight key points
key_points_x = [-2, -1, 0, 1, 2]
key_points_y = [1 + 1 * b1(x) for x in key_points_x]
ax.scatter(key_points_x, key_points_y, color='red', s=100, zorder=5, label='Key points')

# Add labels for key points
for x, y in zip(key_points_x, key_points_y):
    ax.annotate(f'({x}, {y})', (x, y), xytext=(5, 5), textcoords='offset points')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y = 1 + b₁(x)', fontsize=12)
ax.set_title('Basis Function: y = 1 + b₁(x)', fontsize=14)
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(0.5, 2.5)

print("Intercepts:")
print("Y-axis: (0,2)")
print("X-axis: Does not cross the x-axis")

print("Slope Values:")
print("At x < 0: y = 1 + 0 = 1")
print("At 0 ≤ x < 1: y = 1 + 1 = 2")
print("At 1 ≤ x < 2: y = 1 + (2-x)")
print("At x ≥ 2: y = 1 + 0 = 1")
```
