= $\frac{\sqrt{\sum(y_i^2-\frac{(x_iy_i)^2}{x_i^2})}}{\sqrt{(n-1)\sum(x_i^2)}}$

$\frac{\hat{\beta}}{SE(\hat{\beta})}$ = $\frac{\sum(x_iy_i)^2}{\sum(x_i^2)}$ $\frac{\sqrt{(n-1)\sum(x_i^2)}}{\sqrt{\sum(y_i^2-\frac{(x_iy_i)^2}{x_i^2})}}$ = $\frac{\sqrt{(n-1)}\sum(x_iy_i)}{\sqrt{(\sum{x_i^2})\sum(y_i^2-\frac{(x_iy_i)^2}{x_i^2})}}$ 

= $\frac{\sqrt{(n-1)}\sum(x_iy_i)}{\sqrt{(\sum{x_i^2})(\sum{y_i^2})-(\sum{x_iy_i})^2}}$
