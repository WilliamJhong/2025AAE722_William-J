```python
#3
np.random.seed(42)
b = np.random.normal(size=5)
meanb = np.mean(b)
sdb = np.std(b)
print("mean=", meanb, "standard deviation=", sdb)
```

```python
#4
x = [1, 2, 3]
y = [4, 5, 6]
fig, ax = plt.subplots()
ax.scatter(x, y, marker='x')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter Plot of X vs Y')
plt.show()
```
